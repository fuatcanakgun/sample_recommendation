```python
import pandas as pd
from datetime import datetime

df_full = pd.read_csv("random_data.csv", index_col=0)
df_full["date"] = pd.to_datetime(df_full["date"])
df_full.info()
```

## Top sellers

```python
## by quantity all times

df = df_full.groupby("item").quantity.sum()
df = pd.DataFrame(df.sort_values(ascending=False)).reset_index()

df["quantity"] = df["quantity"].astype(int)
df = df[df["quantity"] > 1]

df
```

```python
## by quantity current year
this_year = datetime.now().strftime("%Y")
df = df_full[df_full["date"] > str(int(this_year)-1)]

df = df.groupby("item").quantity.sum()
df = pd.DataFrame(df.sort_values(ascending=False)).reset_index()

df["quantity"] = df["quantity"].astype(int)
df = df[df["quantity"] > 1]

df
```

```python
## by billings all time
df = pd.DataFrame(df_full.groupby(["item"]).transaction_id.count())
df = df.sort_values("transaction_id", ascending=False).reset_index().rename({"transaction_id":"count"}, axis=1)

df["count"] = df["count"].astype(int)
df = df[df["count"] > 1]

df
```

```python
## by billings current year
this_year = datetime.now().strftime("%Y")
df = df_full[df_full["date"] > str(int(this_year)-1)]

df = pd.DataFrame(df.groupby(["item"]).transaction_id.count())
df = df.sort_values("transaction_id", ascending=False).reset_index().rename({"transaction_id":"count"}, axis=1)

df["count"] = df["count"].astype(int)
df = df[df["count"] > 1]

df
```
